```python
#Libraries#
###########
import os
import numpy as np
import matplotlib.pyplot as plt

#for loading and visualizing audio files
import librosa
import librosa.display

#to play audio
import IPython.display as ipd
```

```python
def load_files(folder_name):
    audio_fpath = folder_name
    files = librosa.util.find_files(audio_fpath, ext=['wav'])
    files = np.asarray(files)
    counter = 1

    for y in files:
        print(y)
        data, sr = librosa.load(y, sr=44100)
        X = librosa.stft(data)
        Xdb = librosa.amplitude_to_db(abs(X))
        librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
        #plt.colorbar()
        plt.set_cmap('viridis')
        plt.ylim((19700, 20300))
        plt.xlim((1.3,2.7))
        plt.axis('off')
        plt.savefig("Images/P/P" + str(counter) + ".png",dpi=300, bbox_inches='tight',pad_inches = 0)
        plt.cla()
        counter+=1
        #plt.show()
        #break
        
```
